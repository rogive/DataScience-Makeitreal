# Ejercicio #3 - Matplotlib 

1- Defina tres listas, `x`, `y1`, y `y2` y llénelas con números enteros. Estos números pueden ser cualquier cosa que usted quiera, pero sería bueno que fueran métricas reales que usted quiera comparar.

```python
from matplotlib import pyplot as plt
```

```python
x = range(5)
y1 = [ 5, 8, 12, 16, 18]
y2 = [ 13, 14, 14, 14.5, 16]
```

2- Grafica `y1` vs `x` y muestra la gráfica.

```python
plt.plot( x, y1, color= 'green')
plt.title("y1 vs x")
plt.show()
```

3- En la misma gráfica, grafica `y2` vs `x` (después de la línea donde graficaste `y1` vs `x`)

```python
plt.plot( x, y2, color= 'blue')
plt.title("y2 vs x")
plt.show()
```

4- Haga que la línea `y1` sea una línea rosada y la línea `y2` una línea gris. Dé a ambas líneas marcadores redondos.

```python
plt.plot( x, y1, color = 'pink', marker = "o")
plt.plot( x, y2, color= 'gray', marker = "o")
plt.show()
```

5- Dale a tu gráfico un título de `"Dos líneas en un gráfico"`, y etiqueta el eje `x` `"Increíble eje X"` y el eje `y` `"Increíble eje Y"`.

```python
plt.plot( x, y1, color = 'pink', marker = "o")
plt.plot( x, y2, color= 'gray', marker = "o")
plt.title("Dos lineas en un gráfico")
plt.xlabel("Increible eje X")
plt.ylabel("Increible eje Y")
plt.show()
```

6- Dale al gráfico una leyenda y ponla en la parte inferior derecha.

```python
plt.plot( x, y1, color = 'pink', marker = "o")
plt.plot( x, y2, color= 'gray', marker = "o")
plt.title("Dos lineas en un gráfico")
plt.xlabel("Increible eje X")
plt.ylabel("Increible eje Y")
plt.legend(["Linea 1", "Linea 2"], loc= 4)
plt.show()
```
