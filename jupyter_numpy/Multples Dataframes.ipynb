```python
#Dados los siguientes dataframes: (speaker notes), usted necesita analizar las ventas
#Cree un nuevo dataframe llamado sales_vs_targets el cual contiene el inner join entre sales y targets
#La empresa quiere saber los meses en que alcanzaron sus objetivos (targets) de ventas (revenue). 
#Para ello seleccione las filas de sales_vs_targets donde revenue es mayor a target. Guardelo en una variable 
#llamada crushing_it
```

```python
import pandas as pd
sales = pd.DataFrame({
  'month': ['January', 'February', 'March', 'April', 'May', 'June'],
  'revenue': [300, 290, 310, 325, 475, 495]
})
targets = pd.DataFrame({
  'month': ['January', 'February', 'March', 'April', 'May', 'June'],
  'target': [310, 270, 300, 350, 475, 500]
})
sales
```

```python
targets
```

```python
sales_vs_targets = pd.merge(sales, targets)
sales_vs_targets
```

```python
crushing_it = sales_vs_targets[sales_vs_targets.revenue > sales_vs_targets.target]
crushing_it
```

```python
men_women = pd.DataFrame({
  'month': ['January', 'February', 'March', 'April', 'May', 'June'],
  'men': [30, 29, 31, 32, 47, 49],
  'women': [35, 35, 29, 28, 50, 45]
})
men_women
```

```python
#Tenemos más datos sobre los dataframes de ventas. Ahora tenemos el numero de ventas por mes por género (hombre o mujer) - speaker notes
#haga un merge sobre los 3 dataframes en uno grande llamado all_data
#La empresa cree que tiene más ventas en los meses donde venden mas camisetas de mujer. Seleccione todas las filas de all_data donde revenue es mayor a target y women es mayor que men. Guarde e imprima en la variable results.
```

```python
all_data = sales_vs_targets.merge(men_women)
all_data
```

```python
all_data[(all_data.revenue) > all_data.target & (all_data.women > all_data.men)]
```
