```python
df3 = pd.DataFrame([
    ['Apple', 1, 'No'],
    ['Milk', 4.2, 'No'],
    ['Paper Towel', 5, 'Yes'],
    ['Ligth Bulbs', 3.7, 'Yes']
], columns=['Item', 'Price', 'Is Taxed?'])
```

```python
df3
```

```python
df3['Price with tax'] = df3.apply(
    lambda row: row['Price'] * 1.19
    if row['Is Taxed?'] == "Yes"
    else row['Price'],
    axis = 1
)
```
