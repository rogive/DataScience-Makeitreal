# Ejercicio #2 - Matplotlib

````python
from matplotlib import pyplot as plt

months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]
````

1- Hemos definido las listas de `months`, `temperature` y `flights_to_hawaii` para ti. Usando el comando de subplot, grafica `months` vs `temperature`,  en la caja de la izquierda de una figura que tiene 1 fila con 2 columnas.

````python
plt.subplot(1, 2, 1)
plt.plot( months, temperature, color= 'green')
plt.title("months vs temperature")

plt.show()
````

2- Grafica `flights_to_hawaii` vs `temperature` en la misma figura, a la derecha de tu primer subplot. Añada el parámetro `o` al final de su llamada a `plt.plot` para convertir el gráfico en un gráfico de dispersión, si lo desea!

````python
plt.subplot(1, 2, 2)
plt.plot( flights_to_hawaii, temperature, "o", color= 'green')
plt.title("flights to hawaii vs temperature")

plt.show()
````
