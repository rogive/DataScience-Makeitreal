```python
import plotly.graph_objects as go
```

```python
stock = 'MSFT'
start2 = datetime(2019, 1, 1)
 
df = web.DataReader(stock, data_source='yahoo', start=start2)
```

```python
df
```

```python
trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'MSFT',
    'showlegend': True
}
```

```python

# Config graph layout
layout = go.Figure(
    data = [trace1],
    layout_title_text="Microsoft(MSFT) Moving Averages"
)
```
