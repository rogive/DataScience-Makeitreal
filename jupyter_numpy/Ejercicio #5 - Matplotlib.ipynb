18- Utilice `fill_between` para sombrear el error por encima y por debajo de la línea que hemos trazado, con un alfa de `0,2`.

````python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
````

````python
plt.plot(months, revenue)
plt.show()
````

````python
import numpy as np
ax=plt.subplot()
y_upper=1.1*np.array(revenue)
y_lower=0.9*np.array(revenue)
plt.plot(months,revenue)
plt.fill_between(months,y_lower,y_upper,alpha=0.2)
ax.set_xticks(months)
ax.set_xticklabels(month_names,rotation=30)
plt.show()
````
